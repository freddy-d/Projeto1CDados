<h2>Introdução:</h2>

<p>Este trabalho tem como objetivo responder uma pergunta:"Há relação entre corrupção e emprego? Isso de alguma forma tem impacto no IDH de certo país?". Através da análise exploratória de três dados, obdtidos em diversos países, no ano de 2005 é possivel chegar a uma conclusão. Para fazer a análise, os três dados, IDH, Taxa de Emprego e Índice de Corrupção, retirados do site "Gapminder", tiveram que ser processados e estudados. O primeiro indicador, IDH, obrigatório para o projeto, se baseia em alguns fatores como expectativa de vida, índice de educação e índice de renda, este varia de 0 a 1, sendo 1 o melhor possível e 0 o pior. O segundo dado escolhido, esse por sua vez de livre escolha, foi a Taxa de Emprego. Essa representa o percentual de pessoas entre 15 e 24 anos que estão empregadas em certo país. O último indicador, também de livre escolha, foi o Índice de Corrupção, cuja a variação é de 0 a 10, sendo 0 um país com muita corrupção e 10 um país sem corrupção.</p>

```python
%matplotlib inline
import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt

def reta(dados_x, dados_y, eixo_x, eixo_y, titulo, lista):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()
 
    fig = plt.figure(figsize=(6, 4))
    plt.scatter(dados_x, dados_y, c='red', alpha=0.8)
    plt.title(titulo)
    plt.xlabel(eixo_x)
    plt.ylabel(eixo_y)

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color='blue')
    plt.tight_layout()
    if len(lista)!=0:
        for i in lista:
            plt.plot(i[0],i[1],i[2])
    plt.show()
    return None

def reta_semponto(dados_x, dados_y, eixo_x, eixo_y, titulo):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()
 
    fig = plt.figure(figsize=(6, 4))
    plt.title(titulo)
    plt.xlabel(eixo_x)
    plt.ylabel(eixo_y)
 
    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color='blue')
    plt.axis([0,1.15*max(dados_x),0,1.15*max(dados_y)])
    plt.tight_layout()
    plt.show()
    return None

def pontos(lista, eixo_x, eixo_y, titulo):
    legenda=[]
    if len(lista)!=0:
        for i in lista:
            plt.plot(i[0],i[1],i[2])
            legenda.append(i[3])
    plt.legend(legenda)
    plt.title(titulo)
    plt.xlabel(eixo_x)
    plt.ylabel(eixo_y)
    plt.show()
    return None
```

```python
idh=pd.read_excel("idh.xlsx")

idh=idh.loc[:,["HDI",2005]].dropna()
idh=idh.rename(columns={2005:"IDH","HDI":"País"})
idh=idh.set_index("País")

ig, axes = plt.subplots(1,2,figsize=(8,6))
idh.plot.box(ax=axes[0])
idh.hist(ax=axes[1])
plt.show()
```

```python
emprego=pd.read_excel("emprego.xlsx")

emprego=emprego.loc[:,["geo",2005]].dropna()
emprego=emprego.rename(columns={2005:"Emprego","geo":"País"})
emprego=emprego.set_index("País")

ig, axes = plt.subplots(1,2,figsize=(8,6))
emprego.plot.box(ax=axes[0])
emprego.hist(ax=axes[1])
plt.show()
```

Ao analisar os dados referente a Taxa de Emprego, é possível notar pelo boxplot, que a mediana fica em torno de 40% (representada pela linha verde), sendo que ela está bem distribuida entre o primeiro e o terceiro quartial, estando quase no meio deles. Estes por sua vez estão quase centralizados no boxplot, sendo que o primeiro quartil está a 9.1 do valor mínimo e o terceiro a 23.4 do valor máximo. Isso representa uma destribuição simétrica dos dados, como também é possível notar atraves da analise do histograma. Outro fator que representa a simetria dos dados é que não há a presença de outliers no boxplot. Essa simetria representa que para este ano, a maioria dos países tinha entre 20% e 40% da sua população, entre 15 e 24 anos, com emprego.

```python
corrupcao=pd.read_excel("corrupcao.xlsx")

corrupcao=corrupcao.loc[:,["geo",2005]].dropna()
corrupcao=corrupcao.rename(columns={2005:"Corrupção","geo":"País"})
corrupcao=corrupcao.set_index("País")

ig, axes = plt.subplots(1, 2,figsize=(8,6))
corrupcao.plot.box(ax=axes[0])
corrupcao.hist(ax=axes[1])
plt.show()
```

Já ao analisar os dados referentes a corrupção, olhando ao histograma primeiramente, imediatamende já é notável que os valores não apresentão uma simetria. Neste caso uma distribuição assimétrica positiva é evidente, ou seja uma quantia dos dados, referente aos países com Índice de Corrupção menor que 5, estão concentradaos a esquerda do gráfico. Isso faz com países que tenham o índice elevado se tornem exceções. É possível perceber isto ao observar o pontos no boxplot, que representão países com a taxa de corrupção acima de 8.75, estes são chamados de outliers ou seja casos considerados de valor extremo, uma anormalidade. O boxplot também aponta uma mediana próxima do 3, o primeiro quartil no 2.5 e o terceiro no 5. 

Em suma, todos esse fatores representão que uma boa parte dos países tem Índice de Corrupção baixo, relembrando que quanto menor o índice maior a quantidade de corrupção no local, mas que também há países com pouca corrupção.

```python
dataframe=corrupcao.join([emprego,idh],how="inner")

dataframe.sort_values(["IDH"], ascending=False).head(10)
```

```python
print("Coeficiente de Correlação Linear: {}".format(dataframe["Corrupção"].corr(dataframe["Emprego"])))

reta(dataframe["Corrupção"],dataframe["Emprego"],"Corrupção","Emprego", "Índice de Corrupção por Taxa de Emprego em 2005",[[2.7,12.4,"go"],[2.5,75.7,"bo"],[9.7,71.6,"yo"],[9.6,39.8,"ko"]])

pontos([[2.7,12.4,"go","Macedonia"],[2.5,75.7,"bo","Madagascar"],[9.7,71.6,"yo","Islândia"],[9.6,39.8,"ko","Finlândia"]],"Corrupção","Emprego","Índice de Corrupção por Taxa de Emprego em 2005")
```

Analisando o gráfico acima e o Coeficiente de Correlação Linear de aproximadamente 0.017, é possível concluir que o Índice de Corrupção e a Taxa de Emprego não possuem uma relação entre si. São quatro casos de combinações, do que pode se ocorrer em um país.

Um país pode ter o índice de corrupção baíxo (ocorre muita corrupção no local) e ter uma pequena porcentagem de pessoas com emprego. Por exemplo, para este primeiro caso, é o ponto destacado em verde, que representa a Macedonia, possui um índice de corrupção de 2.7 e uma taxa de pessoas empregadas perto dos 12%. Esse caso deve occorer por conta de algúns fatores no país, entre eles a corrupção 

O segundo caso é quando o índice de corrupção é alto, quase não se tem corrupção no país, e a taxa de emprego continua sendo baixa. A Finlândia, representada pelo pronto preto, aponta uma taxa de emprego próximo a 40% e corrupção de 9.6, é um bom exemplo para este caso.  

O oposto em relação ao percentual de empregados também ocorre, há países com muitas pessoas empregadas acima de 70% e com muita corrupção, aproximadamente 2.5, nesse caso Madagascar representado pelo ponto azul no gráfico é um exemplo para este terceiro caso. 

Mas também há países como a Islândia, representada por amarelo com 9.7 de índice de corrupção e 71.6% de pessoas empregadas, ou seja há pouca corrupção e muitos empregados, repesenta o quarto e útlimo caso do que pode ocorrer. Este caso é o que seria considerado o melhor para um país, pouca corrupção e muitos empregados, porém como veremos a seguir isso não necessariamente reflete no IDH.

Em suma, e respondendo a primeira pergunta, não há uma relação direta entre a Taxa de Emprego e Índice de Corrupção, cada país teria de ser analisado individualmente para determinar como se estabelce a relação entre os indicadores.

```python
cor=dataframe["IDH"].corr(dataframe["Emprego"])
lista=[]
reta(dataframe["IDH"],dataframe["Emprego"],"IDH","Emprego", "IDH por Taxa de Emprego em 2005",lista)
print("Coeficiente de Correlação Linear: {}".format(cor))
```

Já neste segundo gráfico de comparação de indicadores, assim como no gráfico acima, não é possivel dizer que há uma relação direta entre emprego e IDH. O Coeficiente de Correlação Linear, próximo de -0.340 aponta para isso. Novamente, como descrito acima, há quatro casos do que se pode ocorrer, alto IDH com muito emprego, alto IDH com pouco emprego, baixo IDH com muito emprego e baixo IDH com pouco emprego. Ou seja teria de ser feita uma análise de cada país para entender de que forma o valor da porcentagem de empregados altera o IDH do mesmo. 

```python
cor=dataframe["IDH"].corr(dataframe["Corrupção"])
lista=[]
reta(dataframe["IDH"],dataframe["Corrupção"],"IDH","Corrupção", "IDH por Índice de Corrupção em 2005",lista)

print("Coeficiente de Correlação Linear: {}".format(cor))
```

Com o Coeficiente de Correlação Linear perto de 0.734, este terceiro gráfico e último gráfico de comparação de indicadores entre IDH e Índice de Corrupção, apesar de ser o útlimo, é o primeiro que realmente aponta uma relação entre os dois indicadores. Isso faz com que, em geral, um país que tenha alto IDH tenha um alto índice de corrupção (são poucos os casos de corrupção que ocorre neste país) e o contrário também é válido.

Somente por esses dois dados não é possível determinar qual dos dois é a causa e qual é a consequencia um em relação ao outro, porém com outros dados e para futuras análises seria algo interresante de se estudar.
