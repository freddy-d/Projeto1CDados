```python
%matplotlib inline
import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
idh=pd.read_excel("idh.xlsx")

idh=idh.loc[:,["HDI",2005]].dropna()
idh=idh.rename(columns={2005:"IDH","HDI":"País"})
idh=idh.set_index("País")

idh.head(5)
```

```python
desemprego=pd.read_excel("emprego.xlsx")

desemprego=desemprego.loc[:,["geo",2005]].dropna()
desemprego=desemprego.rename(columns={2005:"Emprego","geo":"País"})
desemprego=desemprego.set_index("País")

desemprego.head(5)
```

```python
corrupcao=pd.read_excel("corrupcao.xlsx")

corrupcao=corrupcao.loc[:,["geo",2005]].dropna()
corrupcao=corrupcao.rename(columns={2005:"Corrupção","geo":"País"})
corrupcao=corrupcao.set_index("País")

corrupcao.head(5)
```

```python
dataframe=corrupcao.join([desemprego,idh],how="inner")

dataframe.sort_values(["IDH"], ascending=False)
```

```python
sns.regplot(dataframe["Corrupção"],dataframe["Emprego"])
plt.show()
```

```python
sns.regplot(dataframe["IDH"],dataframe["Corrupção"])
plt.show()
```

```python
sns.regplot(dataframe["IDH"],dataframe["Emprego"])
plt.show()
```
